After observing some of the data ponts it seems we have some missing values we need to attend to : gearbox, fueltype, vehicletype, model and not repaireddamage and the last seen is not identified as a datetime column

```python
# Handle Column Names

current_names = autos.columns
print(current_names)

new_names = {'dateCrawled' : 'date_crawled',
             'offerType' : 'offer_type',
             'vehicleType' : 'vehicle_type',
             'powerPS' : 'power_ps',
             'nrOfPictures' : 'nr_of_pictures',
             'postalCode' : 'postal_code',
             'yearOfRegistration': 'registration_year', 
             'monthOfRegistration' : 'registration_month',
             'notRepairedDamage': 'unrepaired_damage',
             'fuelType' : 'fuel_type',
             'lastSeen' : 'last_seen',
             'dateCreated': 'date_created'}
autos.rename(columns=new_names, inplace=True)
autos.head()
```

```python
autos.describe(include='all')
```

```python
# change odometer to integer by removing extra charcters
autos['odometer'] = autos['odometer'].str.replace('km','')
autos['odometer'] = autos['odometer'].str.replace(',','')

autos['odometer'] = autos['odometer'].astype(int)
autos['odometer'].unique()

```

```python
autos.rename(columns={'odometer' : 'odometer_km'} , inplace = True)
```

```python
# clean the price column
autos['price'] = autos['price'].str.replace("$",'')
autos['price'] = autos['price'].str.replace(",",'')
autos['price'] = autos['price'].astype(int)
```

Remove rows with price = 0 and price over 1M

```python

autos['price'].unique().shape
autos['price'].describe()
autos['price'].value_counts()
autos = autos[autos.price < 1000000]
autos = autos[autos.price != 0]
autos['price'].max()

```

```python
autos['odometer_km'].describe()
autos['odometer_km'].value_counts()
```

Now we will move into cleaning the date fields and understanding their content. We will start with the date crawled which was added by the crawler. Looks like the period is between March 5th and April 7th of 2016

```python
autos['date_crawled'].str[:10].value_counts(normalize=True,dropna=False).sort_index()
```

```python
# now we will do the same for date_created which arrived from the website 
# and we can see that the period is from June 11th 2015 to April 7th 2016

import datetime as dt
autos['date_created'] = pd.to_datetime(autos['date_created'].str[:10])
autos['date_created'].value_counts(normalize=True,dropna=False).sort_index()
```

```python
# in the last seen column which was created by the crawler 
autos['last_seen'].str[:10].value_counts(normalize=True,dropna=False).sort_index()
```

```python
# analyze registration year remove rows below 1900 and above 2016 and we can see
# most cars were registered in the mid 2000
autos=autos[autos['registration_year'].between(1900,2016)]
autos['registration_year'].value_counts().sort_index()
autos['registration_year'].describe()
```
